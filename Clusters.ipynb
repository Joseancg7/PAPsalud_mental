```python
import main as mn
import data as dt 
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data_c = mn.data_c
data_c.Clusters = data_c.Clusters.astype(str)
```

```python
fechas = dt.fechas
fechas['Fecha'] = pd.to_datetime(fechas['Fecha'], format="%d/%m/%Y")

columna_match = []

for i in range(len(data_c)):
        if data_c.iloc[i].name in fechas['Fecha'].to_list():
            columna_match.append('match')
        else:
            columna_match.append('-')


data_c['Fecha_importante'] = columna_match
```

```python
fig=go.Figure((go.Scatter(x=data_c.index, y=data_c['Frecuencia.Llamada075'],
                         mode='lines', name= 'Frecuencia Llamada 075', line_color='#ffe476')))

fig2 = px.scatter(data_c, x=data_c.index,
                          y='Frecuencia.Llamada075', color='Clusters', symbol='Fecha_importante', 
                  labels={"Clusters": "Cluster", "Fecha_importante": "Fecha importante"})

fig2['data'][0]['showlegend']=True




fig.add_traces(fig2.data)


fig.update_layout(legend_title="Clusters", title='Agrupamiento de Frecuencia de llamada 075 y Fechas importantes')


fig.show()
```
