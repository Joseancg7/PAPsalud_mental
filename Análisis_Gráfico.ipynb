## Gráficos de análisis para variables de interés

```python
from scipy import stats 
from matplotlib import pyplot as plt
%matplotlib inline

#GRAFICOS DE VARIABLES DE INTERES
```

```python
#Grafica llamadas por dia
plt.figure(figsize=(20, 10))
plt.plot(data['Fecha'], data['Frecuencia.Llamada075'], '-k', lw=2, label='Llamadas al 075 por dia')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia de llamadas al 075')
plt.show()
```

```python
#¿Existe un efecto acumulativo en el tiempo de estar en Pandemia por SARS-COVID-19?
#Grafica llamadas acumuladas
plt.figure(figsize=(20, 10))
plt.plot(data['Fecha'], data['Frecuencia.Llamada075'].cumsum(), '-k', lw=2, label='Llamadas al 075 acumulado')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia de llamadas al 075')
plt.show()
```

### Conclusiones e ideas para comunicar

- Entre marzo y julio 2021 vemos un gran cambio a la baja en distintas variables como muertes, nuevos casos y hospitalizados. Mismo periodo en el cual la campaña de vacunación comienza a aplicarse de forma masiva a adultos mayores (sector más vulnerable). En este mismo periodo la frecuencia de llamadas presenta un movimiento lateral sin grandes cambios por día.


- A finales de abril 2020 se decalara la fase 3 de la emergencia sanitaria con nuevas medidas para la vida cotidiana (cambio drástico), para mayo podemos ver como se disparan a la alza las frecuencias de llamadas al 075 y de la misma forma una primera ola de altos números de contagios, hospitalizaciones y defunciones en el país y Jalisco con movimientos muy parecidos.


- A pesar del crecimiento exponencial en vacunas aplicadas, vemos una nueva ola de contagios en julio-agosto 2021 (posiblemente debido al regreso de varias actividades como escuelas de forma presencial). Sin embargo, los nuevos contagio, defunciones y hospitalizaciones no se acercan a los máximos pasados y en cuestión de un mes vuelven a presentar un movimiento importante a la baja. Podemos confirmar los buenos resultados de las vacunas aplicadas a pesar del regreso paulatino a la normalidad en trabajos, escuelas y eventos.


- Entre enero y febrero 2021 vemos una relación entre un pico de contagio en Jalisco, hospitalizados en el estado y la frecuencia de llamadas al 075, periodo de mayor miedo y probablemente circunstacias para buscar ayuda.


- Si comparamos las gráficas de nuevos casos por día con las defunciones por día (a nivel estatal o nacional) podemos ver dos grandes picos (febrero y agosto 2021) que a la vez son opuestos. Es decir, para los nuevos contagios son mayores los números en la segunda oleada (agosto 2021) mientras que en las muertes fue mayor en el periodo de (febrero 2021). Podemos intuir que a pesar de un mayor número de contagios ya avanzada la vacunación, las muertes disminuyeron gracias a la protección que esta otorga.
