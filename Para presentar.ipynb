```python
import numpy as np
import pandas as pd
from scipy import stats 
from matplotlib import pyplot as plt
%matplotlib inline

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#Importando datos
data = pd.read_excel ('Archivo_1_Datos_TS.xlsx')
data.head()
```

```python
#fines de semana merman los datos por falta de registros o nulos.
data = data[data['Día']!=1]
data = data[data['Día']!=7]
```

#### Gráficas de datos públicos en Jalisco para su mejor visualización y análisis de sus comportamientos:

```python
fig = make_subplots(rows=3, cols=1, subplot_titles=("Contagios","Defunciones","Hospitalizados"))

fig.add_trace(
    go.Scatter(x=list(data['Fecha']), y=list(data['confirmado.estatal']),
                line = dict(color='red', width=6)),
                row=1, col=1)
fig.update_layout(xaxis_title="Fecha")
    
fig.add_trace(
    go.Scatter(x=list(data['Fecha']), y=list(data['defuncion.estatal']),
                line = dict(color='blue', width=6)),
                row=2, col=1)
fig.update_layout(xaxis_title="Fecha")

fig.add_trace(
    go.Scatter(x=list(data['Fecha']), y=list(data['H.total.paciente.positivo.estatal']),
                line = dict(color='purple', width=6)),
                row=3, col=1)
fig.update_layout(xaxis_title="Fecha")

fig.update_layout(height=700, width=900, title_text="Efectividad de Vacunación en Jalisco",
                  title_font_size=28)
```
